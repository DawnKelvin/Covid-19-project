```python
# Remove unnecessary columns:The project will analyze cases, deaths, recoveries, and vaccinations across countries and time.
df.drop(['iso_code',
       'new_cases_smoothed',
       'new_deaths_smoothed', 'new_cases_smoothed_per_million',
       'new_deaths_smoothed_per_million', 'reproduction_rate',
       'weekly_hosp_admissions_per_million',
       'total_tests_per_thousand', 'new_tests_per_thousand',
       'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
       'positive_rate', 'tests_per_case', 'tests_units', 'new_vaccinations_smoothed',
       'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred',
       'new_vaccinations_smoothed_per_million',
       'new_people_vaccinated_smoothed',
       'new_people_vaccinated_smoothed_per_hundred', 'stringency_index',
       'gdp_per_capita', 'extreme_poverty', 'cardiovasc_death_rate',
       'diabetes_prevalence', 'female_smokers', 'male_smokers',
       'handwashing_facilities', 'hospital_beds_per_thousand',
       'life_expectancy', 'human_development_index',
       'excess_mortality_cumulative_absolute', 'excess_mortality_cumulative',
       'excess_mortality', 'excess_mortality_cumulative_per_million'], axis = 1, inplace = True)
```

The date is formatted as an object while it should actually be a datetime data type. This has to be modified!

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
df.describe(include = 'all')
```

```python
# Handle missing values
# first visualize missing values

df.isnull().sum()
```

```python
print(df.isnull().sum())
```

```python
# Fill the null values using value 0
df.fillna(0, inplace = True)
```

```python
# filter the dataframe to only include columns of interests and include continent, country, icu patients, age older than 65
cols_of_interest = ['continent', 'location', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations','icu_patients','aged_65_older']
df2 = df[cols_of_interest].dropna()
```

```python
df2
```

```python
# check the unique location names
df['location'].unique()
```

```python
# Top continents by vaccination
top_vax = df2.groupby('continent')['total_vaccinations'].max().sort_values(ascending=False).head(10)
print("\nTop continents by total vaccinations:\n", top_vax)
```

```python
# Top countries by vaccinations
pd.set_option('display.max_columns', 100)
country_vax_ = df2.groupby('location')['total_vaccinations'].max().sort_values(ascending=False).head(50)
print("\nTop 10 countries by total vaccinations:\n", country_vax_)
```

```python
# Plot top 10 countries by total vaccinations
plt.figure(figsize=(10, 6))
country_vax_.plot(kind='barh', color='mediumseagreen')
plt.xlabel('Total Vaccinations')
plt.title('Top 10 Countries by Total COVID-19 Vaccinations')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Plot top continents by total vaccinations
plt.figure(figsize=(10, 6))
top_vax.plot(kind='barh', color='mediumseagreen')
plt.xlabel('Total Vaccinations')
plt.title('Top Continents by Total COVID-19 Vaccinations')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Top daily new cases by country
top_cases = df2.groupby('location')['new_cases'].max().sort_values(ascending=False).head(50)
print("\nTop continents by New Cases:\n", top_cases)
```

```python
# plot
# Plot top countries by total new cases
plt.figure(figsize=(10, 6))
top_cases.plot(kind='barh', color='mediumseagreen')
plt.xlabel('New cases')
plt.title('Top 10 Countries by Total COVID-19 new cases')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Top deaths by continent
top_deaths = df2.groupby('continent')['total_deaths'].max().sort_values(ascending=False).head(10)
print("\nTop continents by Total deaths:\n", top_deaths)
```

```python
# plot
# Plot top continents by total deaths
plt.figure(figsize=(10, 6))
top_deaths.plot(kind='barh', color='mediumseagreen')
plt.xlabel('Total Deaths')
plt.title('Top Continents by Total COVID-19 total deaths')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Top deaths by country
country_deaths = df2.groupby('location')['total_deaths'].max().sort_values(ascending=False).head(50)
print("\nTop countries by Total deaths:\n", country_deaths)
```

```python
# plot
# Plot top countries by total deaths
plt.figure(figsize=(10, 6))
country_deaths.plot(kind='barh', color='mediumseagreen')
plt.xlabel('Total Deaths')
plt.title('Top Countries by Total COVID-19 total deaths')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
